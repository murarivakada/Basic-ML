#Tokyo olympics dataset


It is one of the dataset used for understanding basics of machine learning.It contains the information about country wise medals won by each country.No.of gold medals,sliver medals,bronze medals by each country

In this tutriol we use "numpy","pandas",matplotlib and seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
olympic = pd.read_csv('olympic_medal_count.csv', encoding='latin-1')
olympic.shape
```

```python
olympic.head()
```

```python
olympic.tail()
```

```python
olympic.isnull()
```

```python
olympic = olympic.drop('Rank',axis=1)

olympic.info()
```

#Visual Exploratory Data Analysis
scatter plot

```python
plt.rcParams['figure.figsize'] = [6, 20]
plt.scatter(olympic['Gold'], olympic['NOCCode'])
plt.show()



```

```python
plt.rcParams['figure.figsize'] = [6, 20]
plt.scatter(olympic['Silver'], olympic['NOCCode'])
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [6, 20]
plt.scatter(olympic['Bronze'], olympic['NOCCode'])
plt.show()
```

From the above three scatter plot graphs we can give range of the no.of gold,sliver,bronze medals won by each country

```python
sns.stripplot(y='NOCCode',x='Total',data=olympic,size=5)
plt.show()
```

In the above scatter plot we can see the Total no.of medals won by each country

```python
olympic2 = olympic.iloc[0:20]
plt.scatter(olympic2['Total'], olympic2['NOCCode'])
plt.title("country wise total  number of medals")
plt.show()
```

```python
sns.boxplot(x=olympic2["Gold"])
plt.show()
```

```python
sns.boxplot(x=olympic2["Total"])
plt.show()
```

From the Above two boxplots for the no.of gold,total no.of medals

```python
olympic=olympic.drop('RankbyTotal',axis=1)
olympic.info()
```

```python
olympic.hist(edgecolor='black',bins = 25, figsize= (12,6))
plt.show()
```

IN the above we can see how the bronze,sliver,gold medals and total numbers are divided

```python
# this bar which shows the bivariate relation between each pair of features
import warnings
warnings.filterwarnings("ignore")
sns.pairplot(data=olympic,hue='NOCCode',size=4)
plt.show()
```

Heat Map is used to find out the correlation between different features in the dataset. High positive or negative value shows that the features have high correlation.This helps us to select the parmeters for machine learning.

```python
sns.barplot(data=olympic2,x="NOCCode",y="Total")
plt.title("TOtal No.of Medals")
```
